```python
#to load the data:

import pickle
with open('features_dataset/features_dataset.pkl', 'rb') as f:
    new_dataset = pickle.load(f)
```

```python
import pandas as pd
import numpy as np
df = pd.DataFrame(np.array(new_dataset).reshape(201,9), columns = list("abcdefghi"))
df.head()
```

```python
def convert_float(string):
    try:
        x=float(x)
        return x
    except:
        return 0.0
```

```python
df['b'] = df.b.apply(lambda x : int(x))
df['c'] = df.c.apply(lambda x : int(x))
df['d'] = df.d.apply(lambda x : int(x))
df['e'] = df.e.apply(lambda x : convert_float(x))
df['f'] = df.f.apply(lambda x : int(x))
df['g'] = df.g.apply(lambda x : int(x))
df['h'] = df.h.apply(lambda x : int(x)) 
df['i'] = df.i.apply(lambda x : float(x))
```

```python
df
```

```python
hashtags = df.a
df = df.drop(["a"],axis=1)
df.head()
```

```python
#Fonction pour spliter la base en train et test, et dire quel sera le label recherché
def simple_split(X):
    to_drop=['g']
    T=X[140:]
    X=X[0:140]  
    return X.drop(to_drop,axis=1),X.g,T.drop(to_drop,axis=1)
#drop pour retourner un nouvel objet avec la colonne concernée en guise de colonne label (ici la colonne 'prime_tot_ttc')
```

```python
%%time
X,y,T=simple_split(df.copy())
```

```python
from sklearn.cross_validation import train_test_split
X0_train,X0_test,y0_train,y0_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
# Mean Absolute Percentage Error
def mape_error(y_true, y_pred): 
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
%%time
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=300, max_depth=7, n_jobs=-1, max_features=None, min_samples_split=5)
rf.fit(X0_train, y0_train)
y_rf = rf.predict(X0_test)
```

```python
print("La moyenne absolue de pourcentage d'erreur est de %s %%") %(mape_error(y0_test, y_rf))
```
